### 1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).

```python
# Create dates for the entirety of 2017 and collect all data for 2017
datastart = '2017-01-01' 
dataend = '2017-12-31'
data = requests.get('https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?api_key='\
                 +API_KEY+'&start_date='+datastart+'&end_date='+dataend)
```

### 3. Calculate what the highest and lowest opening prices were for the stock in this period.

```python
# Find the high and low opening by calling max and min, respectively, on the Open column
high_open = max(x for x in dictafx['Open'] if x != None)
low_open = min(x for x in dictafx['Open'] if x != None)

print('The highest opening price in 2017 was '+str(high_open))
print('The lowest opening price in 2017 was '+str(low_open))
```

### 7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
# Median_vol function sorts values, finds the size of the values, and creates an index from the length to then
# find the appropriate median depending on whether there are an even or odd number of values.

def median_vol(values):
    sort_list = sorted(values, reverse=False)
    len_list = len(sort_list)
    index = int(((len_list - 1) / 2))
    
    if len_list % 2 == 1:
        return sort_list[index]
    else:
        return (sort_list[index] + sort_list[index + 1]) / 2
    
print('The median trading volume for 2017 was ' + str(median_vol(dictafx['Traded Volume'])))        
```
